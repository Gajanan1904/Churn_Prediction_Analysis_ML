```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv(r"C:\Users\omsad\Downloads\Customer_Data.csv")

print(df.head())
```

```python
# 2. Convert columns to numbers
df['Tenure_in_Months'] = pd.to_numeric(df['Tenure_in_Months'], errors='coerce')
df['Monthly_Charge'] = pd.to_numeric(df['Monthly_Charge'], errors='coerce')
df['Total_Revenue'] = pd.to_numeric(df['Total_Revenue'], errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
```

```python
# 3. Create churn column
df['Churned'] = np.where(df['Customer_Status'] == 'Churned', 1, 0)
```

```python
# 4. Filter age 25–60
df = df[(df['Age'] >= 25) & (df['Age'] <= 60)]
```

```python
# 5. Basic info
print("Total Customers:", len(df))
print("Churn Rate:", df['Churned'].mean() * 100)
print("Average Monthly Charge:", df['Monthly_Charge'].mean())
print("Average Tenure:", df['Tenure_in_Months'].mean())
```

```python
# 6. Customers by state
df['State'].value_counts().plot(kind='bar')
plt.title("Customers by State")
plt.show()
```

```python
# 7. Revenue distribution
sns.histplot(df['Total_Revenue'])
plt.title("Revenue Distribution")
plt.show()
```

```python
# 8. Monthly charge vs tenure
sns.scatterplot(data=df, x='Tenure_in_Months', y='Monthly_Charge')
plt.title("Monthly Charge vs Tenure")
plt.show()
```

```python
# 9. Charge vs churn
sns.boxplot(data=df, x='Customer_Status', y='Monthly_Charge')
plt.title("Monthly Charge by Customer Status")
plt.show()
```
